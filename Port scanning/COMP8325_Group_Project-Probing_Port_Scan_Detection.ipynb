```python
import pandas as pd
import numpy as np

normal_data = pd.read_csv("normal_dataset.csv")
normal_data['Label'] = 1

malicious_data = pd.read_csv("malicious_dataset.csv")
malicious_data['Label'] = -1

merged_data = pd.concat([normal_data, malicious_data], axis=0)
merged_data
```

## Pre-Processing
-Fill NaN with 0 \
-Convet hex values \
-remove source ip, distination ip, and frame infomation time because it's not relevent \
-remove tcp.urgent_pointer and frame_info.encap_type have all the same value 

```python
fields = ['eth.type', 'ip.id', 'ip.flags', 'ip.checksum', 'ip.dsfield', 'tcp.flags', 'tcp.checksum']
for field in fields:
    merged_data[field] = merged_data[field].apply(lambda x: int(str(x), 16))
    
merged_data = merged_data.fillna(0)
merged_data = merged_data.drop(columns=['frame_info.time','ip.src','ip.dst','frame_info.encap_type',
                                        'tcp.urgent_pointer'.'frame_info.encap_type'])
merged_data
```

## Split test and train data

```python
from sklearn.model_selection import train_test_split
X_train, X_test = train_test_split(merged_data, test_size=0.2, random_state=0)
```

## Isolation forest

```python
# Task1: Anomaly Detection Method 1 For Probing Dataset
from sklearn.ensemble import IsolationForest
from sklearn.metrics import accuracy_score

model = IsolationForest(contamination='auto', random_state=0)
model.fit(X_train.drop(columns=['Label']))

y_pred = model.predict(X_test.drop(columns=['Label']))


accuracy = accuracy_score(X_test['Label'], y_pred )

print("Accuracy:", accuracy)
```
